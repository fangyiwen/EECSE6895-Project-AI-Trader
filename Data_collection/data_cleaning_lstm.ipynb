```python
import numpy as np
import pandas as pd
from yahoo_fin.stock_info import get_data, get_quote_table, get_stats, get_quote_data, get_market_status, get_quote_data
import csv
```

```python
# ticker_list = ['600519.SS', '601398.SS', '600036.SS', '601288.SS', '601318.SS']
ticker_list = ['600519.SS']

historical_datas = {}
for ticker in ticker_list:
    historical_datas[ticker] = get_data(ticker, start_date = None, end_date = None, index_as_date = False, interval = "1d")

for ticker in historical_datas:
    stock = historical_datas[ticker]
    stock.dropna(axis=0, how='any', inplace=True)
    print(stock[len(stock)-7:len(stock)])
    stock.drop(labels=['ticker', 'adjclose', 'date', 'volume'],axis=1, inplace=True)
    order = ['open', 'high', 'low', 'close']
    stock = stock[order]
    stock.to_csv('./output/' + ticker + '.csv', index=None)
```
