```python
# https://algotrading101.com/learn/yahoo-finance-api-guide/
# http://theautomatic.net/yahoo_fin-documentation/
# yahoo_fin

from yahoo_fin.stock_info import get_data, get_quote_table, get_stats, get_quote_data, get_market_status, get_quote_data
import csv
```

```python
# Load CSV data
#  SSE Composite Index (000001.SS), Shenzhen Component (399001.SZ)
ticker_list = ['000001.SS', '399001.SZ']

sse_csv = csv.reader(open('./data/SSE_A-shares Main Board.csv', 'r', encoding='utf-8'))
count = -1
for row in sse_csv:
    count += 1
    if count == 0:
        continue
    if count > 5:
        break
    ticker_list.append(row[5])

szse_csv = csv.reader(open('./data/SZSE_A-shares Main Board.csv', 'r', encoding='utf-8'))
count = -1
for row in szse_csv:
    count += 1
    if count == 0:
        continue
    if count > 5:
        break
    ticker_list.append(row[19])

print(ticker_list)
print(len(ticker_list))
```

```python
historical_datas = {}
for ticker in ticker_list:
    historical_datas[ticker] = get_data(ticker, start_date = None, end_date = None, index_as_date = False, interval = "1d")
```

```python
historical_datas['600000.SS']
```

```python
for ticker in historical_datas:
    historical_datas[ticker].to_csv('./output/' + ticker + '.csv')
```
